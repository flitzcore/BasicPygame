```python
import pygame
import math

pygame.init()
screen = pygame.display.set_mode((800, 600))
clock = pygame.time.Clock()

# Triangle dimensions
BASE = 30   # Side C (base)
SIDE = 150   # Equal sides A and B

# Rotation angle
angle = 0

# Center of rotation: midpoint of base
pivot = (400, 300)  # screen center

def rotate_point(point, pivot, angle_deg):
    angle_rad = math.radians(angle_deg)
    ox, oy = pivot
    px, py = point
    qx = ox + math.cos(angle_rad) * (px - ox) - math.sin(angle_rad) * (py - oy)
    qy = oy + math.sin(angle_rad) * (px - ox) + math.cos(angle_rad) * (py - oy)
    return (qx, qy)

def get_triangle_points(pivot, base_length, side_length):
    half_base = base_length / 2

    # Base points (horizontal)
    p1 = (pivot[0] - half_base, pivot[1])
    p2 = (pivot[0] + half_base, pivot[1])

    # Height of the triangle using Pythagoras
    height = math.sqrt(side_length**2 - half_base**2)

    # Top point
    p3 = (pivot[0], pivot[1] - height)

    return [p1, p2, p3]

running = True
while running:
    dt = clock.tick(60) / 1000

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()
    if keys[pygame.K_a]:
        angle += 180 * dt
    if keys[pygame.K_d]:
        angle -= 180 * dt

    # Compute triangle points and rotate them
    triangle = get_triangle_points(pivot, BASE, SIDE)
    rotated_triangle = [rotate_point(p, pivot, angle) for p in triangle]

    # Drawing
    screen.fill((255, 255, 255))
    pygame.draw.polygon(screen, (0, 120, 255), rotated_triangle)
    pygame.draw.circle(screen, (255, 0, 0), (int(pivot[0]), int(pivot[1])), 5)  # pivot

    pygame.display.flip()

pygame.quit()
```
